```python
# vamos a escalar o normalizar nuestros datos a una escala entre -1 y 1
from sklearn.preprocessing import StandardScaler,normalize
# Standardize data (escalado)
scaler = StandardScaler() 
scaled_df = scaler.fit_transform(df) 
  
# Normalizing the Data (normalizado)
normalized_df = normalize(scaled_df) 
  
# Converting the numpy array into a pandas DataFrame 
normalized_df = pd.DataFrame(normalized_df,columns=df.columns) 
```

## Usando K-Means

```python
from sklearn.cluster import KMeans
```

```python
# realizamos nuestro cluster (hemos escogido indice 2) con nuestros datos normalizados
n_clusters = 2
km = KMeans(n_clusters=n_clusters)
km.fit(normalized_df);
```

```python
# ANTES DE REALIZAR EL GRÁFICO

# Bueno, antes de graficar necesito, de la paquetería
# sklearn.decomposision, importar a PCA. PCA va a ser un método que vamos a
# ocupar, recordando que el conjunto de datos tenía 17 variables, las 17
# variables si las quisiéramos graficar sería un espacio de 17 dimensiones,
# no creo que sea fácil de entender 17 dimensiones.
# Para eso vamos a ocupar un método de decomposición en el que las 17 variables
# diferentes las va a intentar emular, o intentar resumir en dos variables. Esas
# dos variables generadas van a ser como que un reflejo de la importancia de las
# 17 variables como tal y son las que vamos a poder graficar. 

from sklearn.decomposition import PCA
pca = PCA(n_components = 2) 

# Nuestras dos variables van a surgir de hacer la transformación de los
# conjuntos normalizados y eso lo que vamos a graficar.

X_principal = pca.fit_transform(normalized_df) 
X_principal = pd.DataFrame(X_principal)
X_principal.columns = ['P1', 'P2'] 
X_principal.head()
```

```python
# Realizo una gráfica tipo scatter

plt.scatter(X_principal['P1'], X_principal['P2'], c = km.predict(normalized_df));

# ¿Cómo sé cuál es el número de clusters que debo usar? El número clusters es nuestro parámetro
# más importante. La primera forma que les vamos a presentar para decidir cuántos
# clusters vamos a usar es un método que se llama el "método del codo", más conocido
# por su nombre en inglés "elbow method".
```

## Elbow method


```python
# Este método parte de la suma de las distancias al cuadrado, y para cada "k" en el rango del 1 al 15
# va a intentar probar cuál sería la suma de distancias para un conjunto que tenga
# un solo cluster: k=1, dos clusters k=2, tres clusters k=3 y así sucesivamente hasta llegar a 15 clusters.

Sum_of_squared_distances = []
for k in range(1,15):
    km = KMeans(n_clusters=k)
    km = km.fit(normalized_df)
    Sum_of_squared_distances.append(km.inertia_)


# Cuando se utiliza un algoritmo de optimización como el descenso de gradiente, hay un error se que intenta 
# minimizar. Si bien k-means no intenta explícitamente minimizar ningún error, todavía tenemos una forma 
# de medir el error. Se llama inertia. La inertia es la suma de distancias cuadradas de muestras a su
# centro de agrupación más cercano. Cuanto más pequeña es la inercia, mejor.
```

```python
plt.plot(Sum_of_squared_distances);
```

Nuestro método se llama método del codo porque si uno pone esta gráfica en donde se hace una especie de codo o una especie de de división en que los datos iban bajando con una velocidad y de la nada después de ese punto empiezan a bajar mucho más rápido, supuestamente los clusters que representen este punto son el número de clusters que vamos a ocupar, pero este es un método que parece ser un poco arbitrario. Muchos investigadores critican mucho que este método es un tanto arbitrario y está sujeto a la preferencia o delimitación del investigador.

Es ambiguo pero podemos determinar que el optimo es de 4 a 6 clusters; al agregar mas clusters no se obtendra ganancia significativa

## Silhouette method

¿Qué alternativas tenemos? Hay otro método más aceptado o menos ambiguo que se llama el método de la silueta. Explicar el método de la silueta como tal se sale de los alcances de este curso, pero eso no significa que no podamos implementarlo.

```python
# nos dice: importando la métrica (porque esto ya es importante, ya es una métrica)
# de la paquetería sklearn.metrics, vamos a importar el score silueta (silhouette_score)

# ¿qué es lo importante que se debe saber? Lo importante es que es un score que se intenta minimizar. 
# A medida que sea más chico es más conveniente para nosotros que un score alto.

from sklearn.metrics import silhouette_score
silhouette_scores = [] 

for n_cluster in range(2, 8):
    silhouette_scores.append( 
        silhouette_score(normalized_df, KMeans(n_clusters = n_cluster).fit_predict(normalized_df))) 
    
# Plotting a bar graph to compare the results 
k = range(2,8)
plt.bar(k, silhouette_scores) 
plt.xlabel('Numbero de clusters', fontsize = 10) 
plt.ylabel('Silhouette Score', fontsize = 10) 
plt.show() 
```

```python
# Tenemos los cluster que van de dos a siete clusters. Nuestro valor mínimo
# que tenemos visualmente es el de tres y el de cuatro clusters. Según el método de
# la silueta usar 3 o 4 clusters separaría bien nuestro conjunto. Nosotros ahorita,
# para respetar ambos métodos, vamos a tomar el 4, porque el 4 coincidía hacer
# con lo que nosotros observamos en el método del codo o del elbow, y el 4 es el
# mínimo del método de la silueta.

n_clusters = 4
km = KMeans(n_clusters=n_clusters)
km.fit(normalized_df);
plt.scatter(X_principal['P1'], X_principal['P2'], c = km.predict(normalized_df));

# Aquí ya tenemos cuatro conjuntos en el que fueron separadas e identificadas
# características de nuestro conjunto de datos. Recuerden que lo que estamos
# observando es una adaptación de las 17 variables diferentes adaptadas a un
# espacio de dos dimensiones. 
```

```python
#Agregamos la etiqueta del cluster a nuestro conjunto de datos original
df["c"] = km.predict(normalized_df)
df.head()

# Aquí ya tenemos de nuevo nuestro conjunto de datos original, ya no el
# normalizado. A cada uno de esos elementos le agregamos una columna "c" que depende
# de qué cluster corresponde.

# Nuestros datos normalizados nos
# sirvieron para calcular en qué cluster iban a pertenecer pero ya no nos van a
# servir de aquí en delante. Simplemente nos va a servir el resultado de haberlos
# normalizado que se lo vamos a agregar a nuestro data frame original. Por eso ven
# que ya estamos de vuelta trabajando con el data frame original, ya no con el
# normalizado. 

# ¿de qué nos sirvió todo esto de los
# k-means, de los clusters y todo? Hasta ahora simplemente el algoritmo le
# designó un número a cada uno de nuestros elementos, pero eso a nosotros
# no nos ha dado ningún dato. Vamos a regresarnos a graficar
# todas las columnas de nuestro data frame, pero separando los datos del cluster 1,
# del cluster 2, del cluster 3 y del cuarto cluster.
```

```python
import seaborn as sns
```

```python
# Para cada columna en nuestro data frame, cada una de las 17 columnas, vamos a graficar el
# valor del data frame pero con respecto a la columna "c". Va a haber un histograma para
# los elementos del cluster que estén identificados con un 0, para los
# elementos del cluster que estén identificados con 1, con 2 y con 3.

for col in df:
    grid= sns.FacetGrid(df, col='c')
    grid.map(plt.hist, col)
```
